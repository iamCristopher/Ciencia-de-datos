```python
from estadistica_univariada import covarianza2, correlacion
```

```python
archivo= open("estrellas_vecindadSolar_all.dat","r")
archivo.readline()
BP_tmp=[]
RP_tmp=[]
met_tmp=[]
for lin in archivo:
    met_tmp.append(float(lin.split()[3]))
    BP_tmp.append(float(lin.split()[6]))
    RP_tmp.append(float(lin.split()[7]))
archivo.close()

bp=[]
rp=[]
met=[]
for v0,v1,v2 in zip(met_tmp,BP_tmp,RP_tmp):
    if (v1>0)&(v2>0):
        met.append(v0)
        bp.append(v1)
        rp.append(v2)
```

```python
import matplotlib.pyplot as plt

fig= plt.figure(1,figsize=(4,4),dpi=120)
ax1= fig.add_subplot(111)
ax1.set_xlim(7,19)
ax1.set_ylim(7,19)
ax1.set_xlabel("BP")
ax1.set_ylabel("RP")
cm= plt.cm.get_cmap("jet")
sc= ax1.scatter(bp, rp, marker=".", ec="none", fc="black", s=1, c=met, cmap=cm, vmin=-0.5, vmax=0.3)
cb=plt.colorbar(sc)
cb.set_label("metalicidad")
plt.show()

print("covarianza =",covarianza2(bp,rp))
print("corelacion =",correlacion(bp,rp))
print(min(bp), max(bp))
print(min(rp), max(rp))
```
