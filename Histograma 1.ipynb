```python
import matplotlib.pyplot as plt
import numpy as np
from estadistica_univariada import desviacion_estandar, rango_intercuartilico
```

```python
archivo= open("omegaCen.dat","r")
archivo.readline()
vhelio=[]
for lin in archivo:
    vhelio.append(float(lin.split()[8]))
archivo.close()

bines= []
ancho= (310+150)/70
for i in range(70):
    borde= -150+ i*ancho
    bines.append(borde)
#bins=np.linspace(-150,310,70)
```

```python
fig= plt.figure(figsize=(8,4),dpi=100)
ax1= fig.add_subplot(111)
ax1.hist(vhelio,bins= bines,color= "gray",alpha=0.7)
plt.show()
```

```python
#bines stutges
binesST= int(np.log2(len(vhelio))+1)
print(binesST)
fig= plt.figure(figsize=(8,4),dpi=100)
ax1= fig.add_subplot(111)
ax1.hist(vhelio,bins= binesST,color= "gray",alpha=0.7)
plt.show()
```

```python
#scott
W= 3.49*desviacion_estandar(vhelio)*(len(vhelio))**(-1/3)
print(W)

bines1= int((150+400)/W)
print(bines1)


fig= plt.figure(figsize=(8,4),dpi=100)
ax1= fig.add_subplot(111)
ax1.hist(vhelio,bins= bines1,color= "gray",alpha=0.7)
plt.show()
```

```python
#Regla de Freeman & Diaconis
AnchoFD= 2*rango_intercuartilico(vhelio)*(len(vhelio))**(-1/3)
bines2= int((150+400)/AnchoFD)

fig= plt.figure(figsize=(8,4),dpi=100)
ax1= fig.add_subplot(111)
ax1.hist(vhelio,bins= bines2,color= "gray",alpha=0.7)
plt.show()
```
